## CD als Reflexionsgitter

<div class="alert alert-block alert-warning">
    Eine gewöhnliche CD kann als Reflexionsgitter angesehen werden. Wiederholen Sie die vorherigen Aufgaben mit den Werten für eine CD. Passen Sie die Schirmbreite entsprechend an. Welche Änderungen ergeben Sich beim Übergang vom Doppelspalt zum Gitter?

```python
#Vorbelegung der einzelnen Parameter
n = 10000
y = np.linspace(-0.5,0.5,n) #Es werden n Punkte auf dem Schirm y berechnet
Imax= 10
L = 0.8
d = 20e-6
b = 2e-6
lamb = 633e-9
N = 2

#Modulation durch Gitter und Einzelspalt
DS=(np.sin((N*np.pi*d*y)/(lamb*L))/(N*np.sin((np.pi*d*y)/(lamb*L))))**2
ES=(np.sin((np.pi*b*y)/(lamb*L))/((np.pi*b*y)/(lamb*L)))**2

#Zusammengesetzte Formel
I = Imax*DS*ES
#I = Imax*DS

#Darstellung des Ergebnisses als Diagramm
diag = aw.XYDiag('Intensität Interferenzmuster', 'Berechnete Werte', 'y', 'I', y, I,'-')

#Darstellung der Umhüllenden
ax = diag.getDiagramm()
ax.plot(y,Imax*ES)

plt.show()
```
